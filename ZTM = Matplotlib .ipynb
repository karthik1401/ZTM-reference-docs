#### There are 3 ways to go about using matplotlib..

#### 1st Way

```python
fig = pyplot.figure() # creates a figure
ax = fig.add_subplot() # creates few Axes
pyplot.show()
```

#### 2nd Way

```python
fig = pyplot.figure()
ax = fig.add_axes([1, 1, 1, 1]) # creates the 4 sides
ax.plot(x, y)
pyplot.show()
```

#### 3rd Way (recommended)

```python
fig,ax = pyplot.subplots()
ax.plot(x, y);
```

#### Bar Plots

```python
# Make a plot from Dictionary
nut_butter_prices = {"Almond Butter" : 10, "Peanut Butter" : 8, "Cashew Butter" : 12}
fig, ax = pyplot.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title = 'Nut Butter Store', ylabel = 'Price (in Rs.)');
```

#### Horizontal Bar Plots

```python
# Horizontal Bar Plot
fig, ax = pyplot.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

#### Histogram

```python
# Create some data
x = numpy.random.randn(1000) # this will create sample of Normal Distribution type.

fig, ax = pyplot.subplots()
ax.hist(x);                 # needs x data only.. used in case of frequency
```

#### SubPlots Option 1

```python
# the Tuple way of declaring the Axes.. 
fig, ((ax1, ax2), (ax3, ax4)) = pyplot.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

# Plot to each different Axes.. 
ax1.plot(x, x/2);
ax2.scatter(numpy.random.random(10), numpy.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.set(ylabel = 'Price (in Rs.)');
ax4.hist(x);
```

#### SubPlot Option 2

```python
# The List Index way of declaring the axes
fig, ax = pyplot.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

# Plotting the different Index of Axes
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(numpy.random.random(10), numpy.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(x);
```

### The above all plots are using the pyplot option of plotting.. Which one to use? pyplot or matplotlib OO method?

* When trying to plot something quickly, use pyplot method. 
* When trying to plot something elaborately, use matplotlib OO method.

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
## Pyplot method
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target');
```

```python
## Matplotlib OO method mixed with pyplot
fig, ax = pyplot.subplots(figsize=(10, 6))

over_50.plot(kind = 'scatter', 
            x = 'age', 
            y = 'chol', 
            c = 'target',
            ax = ax);

ax.set_xlim((45, 100));
```

```python
## Matplotlib OO method
fig, ax = pyplot.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x = over_50['age'], 
                    y = over_50['chol'],
                     c = over_50['target'])

# Customize the plot
ax.set(title = 'Heart Disease and Cholestrol',
      xlabel = 'Age',
      ylabel = 'Cholestrol')

# Add legends to the plot
ax.legend(*scatter.legend_elements(), title = 'Target')

# Add a horizontal line over the mean 
ax.axhline(over_50['chol'].mean(), linestyle = '--');
```

```python
# Creating two subplots. 

# Creating a figure
fig, (ax0, ax1)  = pyplot.subplots(nrows = 2, ncols = 1, figsize=(10, 10), sharex = True)

# Adding data to plot
scatter0 = ax0.scatter(x = over_50['age'], 
                    y = over_50['chol'],
                    c = over_50['target'])

scatter1 = ax1.scatter(x = over_50['age'], 
                    y = over_50['thalach'],
                    c = over_50['target'])

# Basic Customizations
ax0.set(title = 'Heart Disease ~ Cholestrol and Age',
      ylabel = 'Cholestrol')

ax1.set(title = 'Heart Disease ~ Maximum HeartRate and Age',
      xlabel = 'Age', 
      ylabel = 'Maximum Heart Rate')

# Adding Legends
ax0.legend(*scatter.legend_elements(), title = 'Target')

ax1.legend(*scatter.legend_elements(), title = 'Target')

# Mean Horizontal line
ax0.axhline(over_50['chol'].mean(), linestyle = '--')

ax1.axhline(over_50['thalach'].mean(), linestyle = '--')

# Add a Title to the entire Figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```

```python
# Customizing the style
pyplot.style.use('seaborn-whitegrid')

fig, ax = pyplot.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x = over_50['age'], 
                    y = over_50['chol'],
                     c = over_50['target'],
                    cmap = 'winter')

# Customize the plot
ax.set(title = 'Heart Disease and Cholestrol',
      xlabel = 'Age',
      ylabel = 'Cholestrol')

# Add legends to the plot
ax.legend(*scatter.legend_elements(), title = 'Target')

# Add a horizontal line over the mean 
ax.axhline(over_50['chol'].mean(), linestyle = '--');
```

```python
# Creating two subplots. 

# Creating a figure
fig, (ax0, ax1)  = pyplot.subplots(nrows = 2, ncols = 1, figsize=(10, 10), sharex = True)

# Adding data to plot
scatter0 = ax0.scatter(x = over_50['age'], 
                    y = over_50['chol'],
                    c = over_50['target'],
                      cmap = 'winter')

scatter1 = ax1.scatter(x = over_50['age'], 
                    y = over_50['thalach'],
                    c = over_50['target'],
                      cmap = 'winter')

# Basic Customizations
ax0.set(title = 'Heart Disease ~ Cholestrol and Age',
      ylabel = 'Cholestrol')

ax1.set(title = 'Heart Disease ~ Maximum HeartRate and Age',
      xlabel = 'Age', 
      ylabel = 'Maximum Heart Rate')

# Setting limits on X & Y Axis
ax0.set_xlim([50, 80])

ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Adding Legends
ax0.legend(*scatter.legend_elements(), title = 'Target')

ax1.legend(*scatter.legend_elements(), title = 'Target')

# Mean Horizontal line
ax0.axhline(over_50['chol'].mean(), linestyle = '--')

ax1.axhline(over_50['thalach'].mean(), linestyle = '--')

# Add a Title to the entire Figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```
